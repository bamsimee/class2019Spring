#### nltk와 re를 불러와준다.

```python
""" Basic Regular Expression Meta-Characters, Including Wildcards, Ranges and Closures
.	        Wildcard, matches any character
^abc	    Matches some pattern abc at the start of a string
abc$	    Matches some pattern abc at the end of a string
[abc]	    Matches one of a set of characters
[^abc]      Matches anything but a set of characters
[A-Z0-9]	Matches one of a range of characters
ed|ing|s	Matches one of the specified strings (disjunction)
*	        Zero or more of previous item, e.g. a*, [a-z]* (also known as Kleene Closure)
+	        One or more of previous item, e.g. a+, [a-z]+
?	        Zero or one of the previous item (i.e. optional), e.g. a?, [a-z]?
{n}	        Exactly n repeats where n is a non-negative integer
{n,}	    At least n repeats
{,n}	    No more than n repeats
{m,n}	    At least m and no more than n repeats
a(b|c)+	    Parentheses that indicate the scope of the operators
"""
```

#### "을 세 번 입력하면 그 안의 문자를 있는 그대로, 줄바꿈이나 여러가지 문장부호, 기호들까지 raw string으로 유지해준다.
#### Basic Regular Expression(문법)을 나열하고 있다. 문자들을 다룰 수 있는 가장 많이 쓰이는 여러 기능들을 보여준다.

```python
dic= nltk.corpus.words.words('en')
print(type(dic))
```

#### nltk패키지 속 corpus폴더의 words폴더의 'en'이라는 파일을 불러서 dic으로 할당한다. 영어사전파일이고 타입은 str들이 들어있는 list이다.

```python
wordlist = [w for w in dic if w.islower()]
print(wordlist[:10])


a=[]
for w in dic:
    if w.islower():
        a.append(w)
print(a[:10])
```

#### islower()은 소문자임을 묻는 함수이다. 소문자이면 해당하고 아니면 해당하지 않을 것이다.
#### list안에 for loop를 넣을 수 있다. if조건을 걸 수도 있다.
#### for loop에 의해 dic 안의 모든 str에 대해 각각 islower()함수를 적용되고, 결과로 소문자인 것만 w로 받아준다. 그 결과는 리스트로 나올 것이고 wordlist로 할당한다. wordlist는 dic과는 달리 소문자로 정제된 list가 될 것이다. wordlist의 0부터 10미만의 index를 출력한다.
#### 변수나 결과 값의 type을 잘 알아야한다. 그래야 어떤 함수를 어떻게 사용할 것인지 정할 수 있다.

##### 밑에 있는 코드는 a를 빈 list로 정의하고 w에 대해 for문을 돌리는데 소문자의 조건을 걸어서 소문자라면 a에 append()함수를 이용하여 쌓는다. 이 코드는 위의 wordlist와 정확히 같은 결과를 만들어낸다.

```python
re.search('abc','ababcde')
```

#### re패키지는 regular expression에 관련된 패키지로 하나의 str에 적용이 된다. list에는 적용이 안된다. 따라서 list에 적용을 하고 싶다면 for문을 사용하여 str에 각각 적용해야한다. 이 함수의 입력은 2개로 첫 번째에 찾는 조건이 오고, 두 번째에 조건이 찾아질 대상이 온다.
#### 이 예시에서는 abcde에서 ab의 조건을 찾게 될 것이고 만약에 찾는 조건이 대상에 없다면 아무것도 입력되지 않고, 조건이 찾아졌다면 span=( , )의 형태로 해당조건이 있는 index(처음)과 (끝+1)의 index값이 나오고, match='abc'라고 찾는 조건이 나온다.

```python
result = [w for w in wordlist if re.search('ed$', w)][:10]
print(result[:10])
```

#### list comprehension이란 list 안에 다양한 조건을 써넣어서 만족시키는 요소들로 list를 쉽게 생성하는 것을 이른다. 
#### 여기서는 list comprehension이 위에서 구한 소문자단어들의 리스트인 wordlist에서 for loop를 돌려서 if뒤에오는 조건에 맞는 요소들을 w로 받아 list를 생성하라는 뜻이 된다. 만약 if부분이 없다면 result는 wordlist와 같아져서 의미가 없을 것이다.
#### str에 대한 조건을 달 때 regular expression을 사용한다. 달러표시는 단어의 끝을 의미한다. 따라서 이 부분의 re.search를 실행하면 'wordlist에서 뽑은 단어인 w에서 'ed'가 마지막부분에 오는 단어'가 해당된다. 이 조건에 해당되는 단어를 w로 받아서 result라는 list를 생성한다. 마지막으로 이 list의 처음부터 10미만까지의 index를 갖는 요소를 출력한다.
#### list comprehension에서 in다음에 오는 부분을 먼저 보면 쉽게 해석할 수 있다. wordlist가 list이고 그 list 속의 요소 수 만큼 for loop가 돌아 각각 if 조건에 맞는 w가 받아진다. wordlist의 제일 처음에는 'a'가 올텐데 'ed'로 끝나지 않으므로 if조건에서 걸러져서 w에 담기지 않을 것이다.

```python
exercise = [w for w in wordlist if re.search('^con',w)]
print(exercise[:5])
```

이런 식으로 'con'으로 시작하는 단어를 찾으라고 하면 찾을 수 있어야 한다.

```python
result = [w for w in wordlist if re.search('^..j..t..$', w)]
print(result[:10])
```

#### ^은 단어의 시작, 달러표시는 단어의 마지막을 의미하고 .은 와이들카드로 아무 character를 의미한다.
#### 따라서 여기에서는 j앞에 두 개의 아무철자가 오고 뒤에 아무 두 개의 철자, t가 오고 뒤에 아무 두 개의 철자가 오는 단어가 w로 받아져서 result의 list를 이룰 것이다. 이 list를 맨 처음부터 10번째(index:9) 단어까지 출력한다.

```python
result = [w for w in wordlist if re.search('^[ghi][mno][jlk][def]$', w)][:10]
print(result[:10])
```

#### [ ]안에 들어있는 철자 중 아무거나 하나가 오는 것을 의미한다. 중복은 안되고 세트 당 하나의 철자만 선택해야한다.
#### 여기서는 첫 철자는 g,h,i중 하나 / 두 번째 철자는 m,n,o중 하나~~~ 이렇게 4개의 철자를 갖는 단어를 찾는다. result값을 10개를 출력하도록 했지만 wordlist안에 조건에 해당하는 단어가 4개 밖에 없기 때문에 4개가 출력된다.

```python
result = [w for w in wordlist if re.search('^[ah]+$', w)][:10]
print(result[:10])
```

#### 시작 표시 뒤에 [ah]가 있으므로 a와 h 중 하나로 단어가 시작한다. +는 앞의 세트나 철자가 1회 이상 반복되는 것을 말한다. 따라서 [ah]세트가 1회 이상 반복되다 끝나는 단어들이 w로 받아질 것이다. 만약 ahah, hahahaha 등이 wordlist에 있었다면 결과 값으로 출력이 됐을 것이다.

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
wsj
```

#### 또 다른 corpus를 불러온다. 알파벳 이전의 기호들부터 시작해서 굉장히 큰 단어장이라고 생각하면 된다. wsj는 Wall Street Journal의 약자이다. 이 속에 나왔던 단어들의 단어장으로 wsj도 list다. re.search가 바로 적용이 안되고 for loop를 돌려서 적용시켜야 한다. 
#### sorted()함수는 정렬함수로 알파벳 순서, 숫자 순서 등으로 사전식으로 list의 내용을 정렬해준다.
#### set는 집합을 의미하며 중복되지 않는 여러 값을 모아놓은 것이다. dict와 마찬가지로 순서와 상관없이 값이 저장된다. set에 str을 넣었을 경우 각 철자들이 따로따로 순서 구분없이 저장된다. 여기서는 이 값을 사전식으로 sorted해주었다.

```python
wordlist = [w for w in wsj if re.search('^[0-9]+\.[0-9]+$', w)]
print(wordlist[:10])
```

#### '-'기호는 범위를 의미한다. 따라서 [0-9]는 0부터 9까지의 숫자 중 하나를 택하라는 의미가 된다. 뒤에 +가 있으므로 반복을 해준 후, .을 입력하려는 것이다. 그러나 regular expression에서 .은 와일드카드를 의미하는 기능이 있으므로 문자그대로의 .을 사용하기 위해서 앞에 \백슬래시를 입력해준다. 백슬래시는 이스케이프문자로 뒤에 오는 기호가 문자 그대로 쓰일 수 있도록 기능한다. 
#### 이 코드에서 찾는 조건에 해당하는 부분은 regular expression에서 굉장히 의미있는 표현으로 [0-9]는 숫자의 전부를 다 포함하고 뒤에 .이 온 후에 [0-9]로 모든 숫자가 여러 번 올 수 있게 했으므로 결론적으로 모든 실수를 표현할 수 있게 된다. 음수는 포함하지 않지만 양의 실수 전체를 찾고 싶다면 이렇게 입력하면 된다.

```python
x=[w for w in wsj if re.search('[0-9]',w)]
print(x[:10])
y=[w for w in wsj if re.search('^[0-9]',w)]
print(y[:10])
z=[w for w in wsj if re.search('[0-9]$',w)]
print(z[:10])
```

만약에 ^와 $를 표시하지 않아서 찾는 조건의 처음과 끝을 지정하지 않았다면 조건에 해당하는 부분이 찾는 대상의 어딘가에 포함되기만 하면 된다.
x는 시작과 끝을 지정하지 않았으니 숫자가 최소 한 개 이상 포함된 단어들이 해당될 것이고 그 중 사전식 순서대로 10번째까지 출력하게 된다. 따라서 '로 시작하는 숫자를 포함하는 단어들이 처음에 출력이 되었다.
y는 숫자로 시작하는 단어를 불러오므로 가장 순서가 빠른 숫자인 0이 앞에오는 숫자들이 출력됐다.
z는 숫자로 끝나는 단어들을 불러왔다.

```python
wordlist = [w for w in wsj if re.search('^[A-Z]+\$$', w)]
print(wordlist[:10])
```

#### ^[A-Z]+를 해석하면 단어의 시작이 1개 이상의 대문자 알파벳으로 시작한다는 것이다. 백슬래시와 달러표시가 있으므로 달러표시 그 자체를 의미하고, 마지막을 의미하는 달러표시가 오므로 마지막이 달러표시로 끝나는 단어를 찾아서 wordlist로 받게된다.

```python
p=[w for w in wsj if re.search('[A-Z0-9]',w)]
print(p[-10:])
q=[w for w in wsj if re.search('[^abc]',w)]
print(q[-10:])
```

[A-Z0-9]는 [abc]가 a 혹은 b 혹은 c가 되듯이 A-Z 혹은 0-9가 된다. 결과적으로 대문자 알파벳이나 숫자가 어느 곳에 하나라도 포함된 단어들이 모두 선택된다. [A-Za-z]를 하면 알파벳 대문자+소문자 중 아무 한 철자를 포함하면 된다는 조건이 된다.
[^abc]처럼 대괄호 안에 ^를 넣으면 대괄호 속 철자를 모두 제외하고 모두 불러온다는 뜻이다.

```python
wordlist = [w for w in wsj if re.search('^[0-9]{4}$', w)]
print(wordlist[:10])
```

#### {n}은 앞에 오는 조건을 몇 번 반복하라는 뜻을 가진다. 여기서는 숫자로 시작하고 숫자가 4번 반복해서 온 후에 끝나는 단어를 의미한다. 결론적으로 4 digits를 찾아서 wordlist에 받게 된다.

```python
wordlist = [w for w in wsj if re.search('^[0-9]+-[a-z]{3,5}$', w)]
print(wordlist[:10])
```

#### 숫자로 시작하고 1회 이상 반복한 후에 문자그대로의 -가 온다. 뒤에 소문자 알파벳이 3~5번 반복한 후에 끝나는 단어를 찾아서 wordlist에 받는다.
#### 결과적으로 처음에 숫자 하나로 시작하는 단어는 없다는 것을 알 수 있다.

```python
wordlist = [w for w in wsj if re.search('^[a-z]{5,}-[a-z]{2,3}-[a-z]{,6}$', w)]
print(wordlist[:10])
```

#### 소문자가 다섯 번 이상 나온 후 -가 등장하고 다시 소문자가 2~3번 반복된 후 -가 등장, 다시 소문자가 6번 이하로 등장하는 단어를 찾아오게 된다.

```python
wordlist = [w for w in wsj if re.search('(ed|ing)$', w)]
print(wordlist[:10])
```

#### 괄호 안의 철자가 |로 구분되어 '혹은'이라는 의미로 해석된다. |로 구분되는 철자덩어리들을 ()로 묶어서 입력하면 된다.
#### 여기서는 ed 혹은 ing가 마지막에 포함되는 단어를 찾아오게되고 (ed|ing|ist)로 세 개, 혹은 더 많은 철자덩어리를 포함하여 검색할 수 있다.

```python
re.findall('ab','ababdddabkabddd')
```

#### re패키지에 포함되어있는 findall함수는 input이 두 개로 첫 번째에는 찾는 조건에 해당하는 str를 넣고, 두 번째에는 조건이 찾아질 대상에 해당하는 str를 넣으면 된다. 앞의 조건을 뒤의 대상에서 모두 찾아서 list로 받게 된다.
#### 예시에서는 대상이 되는 ababdddabkabddd 속에 ab가 총 4개 들어있으므로 4개의 ab를 찾아 list로 받게 된다.

```python
[vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word)]
```

#### 먼저 for loop로 wsj리스트의 각 단어들을 word로 받는다. 그 다음 그 word에 re.findall함수를 돌린다. 각 word단어 안에서 모음이 2개 이상 반복되면 그 모음의 list를 vs로 각각 받는다. 결론적으로 vs에는 wsj에 있는 단어들 중 모음을 2개 이상 포함하고 있는 단어들의 모음들을 list로 받게 된다. wsj list의 개수만큼 for loop가 돌 것이다.
#### wsj라는 사전 속에서 두 개 이상의 모음컬렉션을 다 뽑아온 list가 만들어진다.

```python
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

#### 위에서 만든 vs리스트에 nltk패키지의 FreqDist함수를 적용시킨 값을 fdist로 받는다. FreqDist에 input으로 들어가는 variable type이 str list라는 것을 알 수 있다. 
#### fdist값에 most_common기능을 이용하여 두 개 이상의 모음컬렉션 list에서 가장 빈도 수가 높은 10개의 요소를 튜플 형태로 (요소, 빈도수)로 결과값을 낸다.

#### 대괄호 안에 ^기호가 들어가 있으므로 모음을 제외한 모든 철자를 의미한다. 자음철자가 하나 온 후 아무 철자가 하나 이상 반복된 후에 자음철자가 오는 부분을 word에서 불러온다. 

```python
result = re.findall('[^aeiou].+?[^aeiou]', word)
print(result)
```

#### ?은 앞에오는 철자가 안 오거나 한 번 오는 것을 의미한다. 여기서는 앞에 오는 기호가 +이므로 +기호가 있거나 없는 것을 의미한다. +기호가 있다면 맨 앞과 뒤에 자음철자가 오고 가운데에 아무 철자가 한 번 이상 반복되는 부분을 list로 받을 것이다. +기호가 없는 경우라면 자음철자 사이에 아무 철자가 한 개 있는 부분이 받아질 것이다.

```python
wsj = sorted(set(nltk.corpus.treebank.words()))
fdist = nltk.FreqDist(vs for word in wsj for vs in re.findall(r'[aeiou]{2,}', word))
fdist.most_common(10)
```

#### nltk속 corpus속 treebank를 불러서 사전식으로 정렬하여 wsj에 받는다. 모음이 2번 이상 반복되는 부분을 wsj list의 각 단어에서 찾아서 vs로 list로 받아 FreqDist함수를 돌려서 각 요소와 빈도수를 튜플형태로 저장한다. most_common기능으로 가장 빈도가 높은 10개의 결과를 불러온다.

```python
regexp = r'^[AEIOUaeiou]+|[AEIOUaeiou]+$|[^AEIOUaeiou]'
def compress(word):
    pieces = re.findall(regexp, word)
    return ''.join(pieces)

english_udhr = nltk.corpus.udhr.words('English-Latin1')
print(nltk.tokenwrap(compress(w) for w in english_udhr[:75]))
```

```python
rotokas_words = nltk.corpus.toolbox.words('rotokas.dic')
cvs = [cv for w in rotokas_words for cv in re.findall(r'[ptksvr][aeiou]', w)]
cfd = nltk.ConditionalFreqDist(cvs) # CFD works pairwise only
cfd.tabulate()
```
