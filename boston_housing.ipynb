```python
# description
# https://www.kaggle.com/c/boston-housing

# TensorFlow and tf.keras
import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Helper libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
(x_train, y_train), (x_test, y_test) = tensorflow.keras.datasets.boston_housing.load_data(path='boston_housing.npz')
```

```python
model = Sequential()
model.add(Dense(10, input_shape=[13], activation='relu'))
model.add(Dense(20))
model.add(Dense(1))
model.compile(optimizer='rmsprop', loss='mse')
model.summary()
```

입력에 13개 출력이 1개. summary는 summary를 보여준다. parameter넣은 것에 14가 나온다. 동그라미가 13갠데 왜 14개지? 훈련해야할 weight가 13갠댕.. y절편이(상수) 붙어서 14개가 된다. 마지막에는 무조건 1인 nod가 붙는다. 
맨 오른쪽에 오는 출력은 nod하나로 만들거니까 model.add(Dense(1)) 여기에 output자리가 1이어야한다. 중간에는 hidden layer를 만드는 거니까 윗줄에는 output을 10개로했다. input은 맨 첨에만 들어가면 되니까 뒤에는 없어도 됨.
hidden layer10개 짜리에다가 Dense안에 activation='relu'를 넣어준다.

얘는 구조가 첫째 column에 node13개, 다음 10개, 다음 20개, 마지막 1개. // 두 번째에 relu넣어준 것. 마지막은 linear(아무것도 안 넣어준 것과 같다) / compile할 때 2가지 경우 생각가능. regression과 classification. regression인 경우에는 optimizer='rmsprop', loss='mse'를 넣어주면 된다. classification할 때는 optimizer='adam', loss='어쩌고'
밑에 결과 값에서 parameter숫자는 뭔가? 전에는 parameter는 화살표 개수라고 했었다. 첫 번째 140은 첫번재 dense에 있는 화살표다. 13개와 10개를 연결해주는 것. 13x10은 130인데 왜 140인가. 첫 번째 노드에 마지막에 1 노드를 넣어줬었다. 얘도 있어서 140개가 되는 것이다. 마지막은 왜 21이 되는가. 20곱하기 1은 20인데 또 앞쪽 노드 마지막에 1을 넣어주니까. 앞 dense를 볼 때는 앞에 노드추가. 뒤 dense를 볼 때도 앞에만 노드추가.

```python
model.fit(x_train, y_train, epochs=1000)
```

```python
model.evaluate(x_test, y_test)
```
