```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
import csv
```

#### urllib라이브러리에서 urlopen함수를 가져온다. bs4라이브러리에서 BeautifulSoup함수도 가져오고, csv도 가져온다.

```python
html=urlopen('https://editorial.rottentomatoes.com/guide/all-dreamworks-animation-movies-ranked/')
bsObject=BeautifulSoup(html, 'html.parser')
```

#### urlopen을 이용하여 웹페이지를 불러온다. 로튼토마토의 dreamworks animation tomatometer지수순위를 가져왔다.
#### BeautifulSoup기능을 이용하여 html을 불러왔다.

```python
bsObject
```

```python
for n in range(1,36):
    div=bsObject.find('div',{'id':'row-index-'+str(n)})
    title= div.select('a')[-1].get_text()
    year=div.find('span').get_text()
    rate=div.next_sibling.next_sibling.next_sibling.get_text()
    print(n,title,year,rate)
```

#### 총 35위까지 영화의 순위와 제목, 년도, 로튼토마토 지수 정보를 불러왔다.
#### 순위 정보가 div태그에 있어서 div 중 속성 id의 값이 row-index-순위 인 태그를 div로 지정했다.
#### div태그 중 a태그에서 가장 마지막에 있는 태그에 제목 정보가 있기 때문에 select(a)[-1]로 찾아준 후 get_text기능을 이용하여 제목 text를 불러온다.
#### 해당 div태그에서 span태그를 찾아서 get_text를 해주면 꺽쇠 밖에 존재하는 년도정보가 나오고 그것을 year에 저장한다.
#### rate정보는 div태그의 다음다음에 위치하기 때문에 return(엔터)를 포함하여 3번의 next_sibling기능을 이용하여 태그를 지정하고 get_text로 text정보만 불러온다.
#### 위의 네 가지 정보를 print한다.

```python
with open('crawling_practice.csv','w') as f:
    writer=csv.writer(f)
    writer.writerow(['rank','title','year','tomatometer'])
    
    for n in range(1,36):
        div=bsObject.find('div',{'id':'row-index-'+str(n)})
        title= div.select('a')[-1].get_text()
        year=div.find('span').get_text()
        y=year[1]+year[2]+year[3]+year[4]
        rate=div.next_sibling.next_sibling.next_sibling.get_text()
        writer.writerow([n,title,y,rate])
```

#### 위에서 구한 정보를 csv파일로 저장하기 위한 코드이다.
#### 제목을 crawling_practice.csv로 두고 w 모듈을 이용하여 write기능을 사용한다. 4개의 정보를 한 줄에 입력할 것이기 때문에 writerow기능을 이용한다.
#### 정보를 구하는 코드를 위와 같이 이용하여 각 요소에 할당하고 writer.writerow기능을 이용하여 4개의 정보를 한 줄에 입력하면 된다. 결과적으로는 맨 위의 카테고리에 해당하는 rank, title, year, tomatometer가 위치하고 밑으로 총 35개의 줄이 입력된다. 각 카테고리에 맞는 column이 네 개 형성된 모습을 가진다.
#### year정보는 태그에서 그대로 text를 불러오면 (년도)의 모습으로 정보가 불러져서 y요소에 다시 number만 올 수 있도록 지정해준 후 write해 주었다.
